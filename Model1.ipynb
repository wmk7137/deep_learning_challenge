## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pandas as pd
import tensorflow as tf
import sklearn as skl
import os

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'])

application_df.head()
```

```python
# Determine the number of unique values in each column.
application_cat = application_df.dtypes[application_df.dtypes == "object"].index.tolist()

application_df[application_cat].nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_values = application_df['APPLICATION_TYPE'].value_counts()

application_values
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
threshold = 500
application_types_to_replace = application_values[application_values < threshold].index

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()

application_df
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_values = application_df['CLASSIFICATION'].value_counts()

classification_values
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
threshold = 1
classification_counts = classification_values[classification_values > threshold]

classification_counts
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
threshold = 1500
classifications_to_replace = classification_values[classification_values < threshold].index

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()

application_df
```

```python
# Convert categorical data to numeric with `pd.get_dummies`application_df_dummies = pd.get_dummies(application_df, drop_first=True)application_df_dummies
application_df_dummies = pd.get_dummies(application_df)
application_df_dummies
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df_dummies.drop(['IS_SUCCESSFUL'], axis=1).values
y = application_df_dummies['IS_SUCCESSFUL'].values
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

number_input_features = X_train.shape[1]
hidden_nodes_layer1 =  19
hidden_nodes_layer2 = 17
hidden_nodes_layer3 =  11
hidden_nodes_layer4 = 9

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="tanh")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden Layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Fourth hidden Layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer4, activation="relu"))


# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))


# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
class SaveWeightsEveryNepochs(tf.keras.callbacks.Callback):
    def __init__(self, save_dir, save_freq=5):
        super(SaveWeightsEveryNepochs, self).__init__()
        self.save_dir = save_dir
        self.save_freq = save_freq
        if not os.path.exists(self.save_dir):
            os.makedirs(self.save_dir)

    def on_epoch_end(self, epoch, logs=None):
        # Save weights every `self.save_freq` epochs
        if (epoch + 1) % self.save_freq == 0:
            file_path = os.path.join(self.save_dir, f'{epoch + 1}.weights.h5')
            self.model.save_weights(file_path)
            print(f'Saved weights at epoch {epoch + 1} to {file_path}')
```

```python
# Define the directory where weights will be saved
save_dir = './untitled_project/trial_Model_1/'

# Create the callback instance
save_weights_callback = SaveWeightsEveryNepochs(save_dir, save_freq=5)
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=50, callbacks=[save_weights_callback], validation_data=(X_test_scaled, y_test))
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
